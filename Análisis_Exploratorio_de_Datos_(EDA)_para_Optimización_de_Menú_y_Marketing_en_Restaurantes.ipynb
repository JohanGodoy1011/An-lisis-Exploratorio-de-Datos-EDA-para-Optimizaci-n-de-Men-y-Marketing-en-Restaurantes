```python
# Mostrar información general del conjunto de datos
data.info()

# Observar las primeras filas del conjunto de datos
data.head()
print("Johan Godoy")
```

```python
# Paso 1: Cargar el dataset
df = pd.read_csv("food_coded.csv")

# Paso 2: Crear una nueva columna con la cantidad de comida reconfortante consumida
df["comfort_food_count"] = df["comfort_food"].apply(lambda x: len(str(x).split(',')))

# Paso 3: Limpiar los datos eliminando filas con valores faltantes
df = df.dropna(subset=["GPA", "comfort_food_count"])

# Paso 4: Convertir el tipo de datos de la columna "GPA" a numérico
df["GPA"] = pd.to_numeric(df["GPA"], errors="coerce")

# Paso 5: Calcular la correlación entre GPA y la cantidad de comida reconfortante consumida
correlation = df["GPA"].corr(df["comfort_food_count"])

print("Correlación entre GPA y Cantidad de Comida Reconfortante Consumida:", correlation)
print("Johan Godoy")
```

La correlación entre el GPA y la cantidad de comida reconfortante consumida es de aproximadamente 0.20. Esto sugiere una correlación positiva débil entre estas dos variables. En otras palabras, existe una leve tendencia de que aquellos que consumen más comida reconfortante puedan tener un GPA ligeramente más alto, pero la relación no es muy fuerte.

Es importante tener en cuenta que la correlación no implica causalidad. Esto significa que aunque haya una correlación entre el consumo de comida reconfortante y el GPA, no podemos concluir que el consumo de comida reconfortante cause un GPA más alto o viceversa. Otros factores podrían estar influyendo en esta relación.

```python
# Función para graficar tasas de conversión
def graficar_tasas_conversion(var_predictora, var_predecir, type='line', order=None):
    x, y = var_predictora, var_predecir

    # Generar agrupaciones (groupby), calcular tasa de conversión (mean),
    # multiplicarla por 100 (mul(100))
    grupo = data.groupby(x)[y].mean().mul(100).rename('tasa_conv').reset_index()

    # Y generar gráfica
    if type == 'line':  # Útil para rangos continuos
        plt.figure(figsize=(10, 6))
        sns.lineplot(x=var_predictora, y='tasa_conv', data=grupo)
        plt.grid()
    elif type == 'bar':  # Útil si los datos están divididos en rangos o son categóricos
        plt.figure(figsize=(14, 6))
        sns.barplot(x=var_predictora, y='tasa_conv', data=grupo, order=order)
        plt.grid()
        plt.xlabel(var_predictora)  # Agregado para etiquetar el eje x
        plt.ylabel('Tasa de Conversión (%)')  # Agregado para etiquetar el eje y
        plt.title(f'Tasa de Conversión por {var_predictora}')  # Agregado para agregar un título
    elif type == 'scatter':  # Útil si los datos están divididos en rangos o son categóricos
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=var_predictora, y='tasa_conv', data=grupo)
        plt.grid()
print("Johan Godoy")
```

```python
# Y graficar las tasas de conversión para la variable "fruit_day" y "Gender_bin"
graficar_tasas_conversion('fruit_day', 'Gender_bin')

# Función para graficar tasas de conversión para variables numéricas
def graficar_tasas_conversion_numerica(var_predictora, var_predecir, type='bar'):
    x, y = var_predictora, var_predecir

    # Generar agrupaciones (groupby), calcular tasa de conversión (mean),
    # multiplicarla por 100 (mul(100))
    grupo = data.groupby(x)[y].mean().rename('tasa_conv').reset_index()

    # Normalizar las tasas de conversión para que sumen 100%
    grupo['tasa_conv'] = grupo['tasa_conv'] / grupo['tasa_conv'].sum() * 100

    # Y generar gráfica
    if type == 'bar':
        plt.figure(figsize=(10, 6))
        sns.barplot(x=var_predictora, y='tasa_conv', data=grupo)
        plt.grid()
    elif type == 'scatter':
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=var_predictora, y='tasa_conv', data=grupo)
        plt.grid()
print("Johan Godoy")
```

```python
# Graficar las tasas de conversión para 'fruit_day'
graficar_tasas_conversion_numerica('fruit_day', 'Gender_bin', type='bar')
plt.xlabel('Gender_bin')
plt.ylabel('Tasa de Conversión (%)')
plt.title('Tasa de Conversión por Gender_bin para fruit_day')
plt.show()

# Graficar las tasas de conversión para 'veggies_day'
graficar_tasas_conversion_numerica('veggies_day', 'Gender_bin', type='bar')
plt.xlabel('Gender_bin')
plt.ylabel('Tasa de Conversión (%)')
plt.title('Tasa de Conversión por Gender_bin para veggies_day')
plt.show()

# Realicemos el mismo análisis pero para la variable 'balance'
graficar_tasas_conversion('veggies_day','Gender_bin',type='bar')
print("Johan Godoy")
```

```python
# Seleccionar variables de interés
categorias_interes = ['breakfast', 'coffee', 'drink', 'nutritional_check']

# Crear subplots para cada variable categórica
fig, axes = plt.subplots(nrows=len(categorias_interes), ncols=1, figsize=(10, 15))

# Iterar sobre las variables y graficar
for i, var in enumerate(categorias_interes):
    # Normalizar los datos para obtener porcentajes
    data_norm = data.groupby([var, 'Gender']).size().reset_index(name='count')
    data_norm['percentage'] = data_norm.groupby('Gender')['count'].transform(lambda x: (x / x.sum()) * 100)

    sns.barplot(x=var, y='percentage', hue='Gender', data=data_norm, ax=axes[i])
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Percentage')
    axes[i].set_title(f'{var} por Género')

plt.tight_layout()
plt.show()
print("Johan Godoy")
```
